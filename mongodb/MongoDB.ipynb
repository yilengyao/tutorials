### find function
`db.<collectionname>.find(query, fields, limit, skip, batchsize, options)` is a MongoDB query method used to retrieve documents from a collection. This method returns all documents that match the provided query criteria.

#### Basic Usage
1. Find All Documents: Retrieve all documents in a collection.
```
db.<collectionname>.find()
```

2. Find Documents with Query: retrieve documents that match specific criteria.
```
db.<collectionname>.find(query)
```
query are key value pairs.

- Find Documents with Specific criteria
```
db.users.find({ age: 25 })
```
Retrieve users with age of 25

- Find Documents with Multiple Criteria
```
db.users.find({ age: 25, status: "A" })
```
Retrieve users with age of 25 and status "A"

- Find Documents with Range Query
```
db.users.find({ age: { $gte: 18, $lte: 30 } })
```
This retieves all documents in the `users` collection where the `age` si between `18` and `30` inclusive.

- Find Documents with an Array field
```
db.posts.find({ tage: "mongodb" })
```
This retrieves all documents in the `posts` collection where the `tags` array contains the value "mongodb".

### Projections
Control which fields are returned in the documents by using a projection. If the field does not exist in the document, the document only returns the key value pair of the existing fields.

1. Return only specific fields
```
db.users.find({ age: 25 }, { name: 1, age: 1, _id: 0 })
```
This retrieves all documents in the `users` collection where the `age` is `25`, but only returns the `name` and `age` fields, excluding the `_id` field.

### Limit and Sort
You can limit the number of documents returned and sort them.

1. Limit Results
```
db.users.find().limit(10)
```
This retrieves only the first `10` documents from the `users` collection.

2. Sort Results
```
db.users.find.sort({ age: -1 })
```
This retrieves all documents from the `users` collection sorted by `age` in descending order.

### Combining Limit and Sort
```
db.users.find().sort({ age: 1}).limit(5)
```
This retrieve the first `5` documents from the `users` collection sorted by `age` in ascending order.


### Pagination
Pagination in MongoDB is sued to retrieve a subset of results form a larger dataset, typically by specifying the number of results to skip and the number of results to return. This is useful for displaying large datasets in smaller, more manageable chunks, such as pages in a web application.

Basic Pagination <br>
You can achieve pagination using the `skip()` and `limit()` methods.
1. `skip(m)`: Skips the first m documents.
2. `limit(n)`: Limits the result set to n documents.

Ex <br>
Assume you have a collection named `users` with many documents, and you want to paginate through them , displaying 10 users per page.
1. Page 1: Skip 0 documents, limit 10.
```
db.users.find().skip(0).limit(10)
```
2. Page 2: Skip 10 documents, limit 10.
```
db.users.find().skip(10).limit(10)
```
3. Page 3: Skip 20 documents, limit 10.
```
db.users.find().skip(10).limit(10)
```

Example Function for Pagination
```
function getPaginatedResults(pageNumber, pageSize) {
    const skip = (pageNumber - 1) * pageSize;
    return db.users.find().sort( age: 1 }).skip(skip).limit(pageSize);
}

// Usage
const pageNumber = 2;
const pageSize = 10;
const result = getPaginatedResults(pageNumber, pageSize);
```

## Indexing
Indexing is crucial for improving the performance of query operations in MongoDB. Indexes supports the efficient execution of queries by reducing the amont of data that MongoDB needs to scan to find the documents that match the query.

Benefits of Indexing
1. Faster Query Performance: Indexes significantly speed up read operations by allowing MongoDB to quickly locate the relevant documents.
2. Efficient Sorting: Indexes can improve the performance of queries that require sorting.
3. Support for Range Queries: Indexes enhance the performance of range queries (e.g., finding documents within a certain range of values).

### Creating Indexes
indexes can be created on one or more fields in a colleciton. The most common types of indexes are single-field indexes and compound indexes.

Single-Field Index
```
db.<collectionname>.createIndex({ fieldname: 1 }) // Ascending order
db.<collectionname>.createIndex({ fieldname: -1 }) // Descending order
```
Example:
```
db.users.createIndex({ age: 1 })
```

Compound index
```
db.<collectionname>.createIndex({ field1: 1, field2: -1 })
```

### Unique Index
Indexes in MongoDB are not unique by default. To ensure that a field is unique in a MongoDB colleciton, you need to create a unique index on that field. Unique indexes prevent duplicate values in the indexed field, ensuring that each value is unique across the collection.

#### Creating a Unique Index
You can create a unique index using the `createIndex` method with the `unique` option set to `true`.

Creating a Unique Index on a Single Field
```
db.users.createIndex({ email: 1 }, { unique: true })
```

Creating a Unique Index on Multipl Fields
```
db.users.createIndex({ firstName: 1, lastName: 1}, { unique: true})
```

If you create an unique index on a field that has duplicates, MongoDB will return an error.

# update
Careful Update: accidental update can cause disaster

Normally, update operation requires two things
1. Find query
2. New data or data that we want to edit.

In Mongo we can do two types of updates
1. Replace entire document: Replacement Approach
2. Use update operators to update specific fields - Target update, Operator Approach.

#### Replacement Approach
- Retrieve entire document into your code level
- Edit document at code level
- Put that updated doc back into database with old document id.

Note: Object id/ Unique Id is immutable, can't update.

Problems with Replacement Approach
1. It requires round trip
- - retrieve document and update and put it back
  - Increases network load
2. Simultaneous Update
- - race conditions

#### Operator Approach
- $set: Set a value of a specific field
```
{$set : fieldname1: value, fieldname2: value}}
```
- $unset: remove specifiec field
```
{$unset : {fieldname1: "", fieldname2: ""}}
```
eg.
```
db.employee.update({ _id: userid }, { $set: { aka: "Iron Man", firstName: "Morgan" }})
``` 


### update
The `db.<collectionname>.update()` method in MongoDB is used to modify existing documents ina a collection. This method can update one or more documents that match the specified filter criteria.
```
db.<collectionname>.update(
    <query>,
    <update>,
    {
        upsert: <boolean>,
        multi: <boolean>,
        writeConcern: <document>
    }
)
```
- query: The selection criteria for the update. It determines which documents to update
- update: The modifications to apply.
- multi: true - enable multiple document updates
- upsert: true - add document when no document match our condition
- writeConcern: The level of acknoledgement requested from MongoDB for write operations.

updating a single document
```
db.employee.update(
    { name: "Alice" },
    { $set: { position: "Manager" }}
)
```

updating multiple documents
```
db.employee.update(
    { department: "Sales" },
    { $set: { status: "Active" }},
    { multi: true }
)
```

upserting a document
```
db.employee.update(
   { name: "Charlie" },
   { $set: { name: "Charlie", position: "Intern" } },
   { upsert: true }
)
```

#### $set and $unset
The `$set` and `$unset` operators are used to update operations to modify the fields of a document. The `$set` operator is used to update the value of a field or add a new field if it does not exist, while the `$unset` operator is used to remove a field from a document.

##### $set
The `$set` operator update the value of a field or adds the field if it does not exist. it is commonly used in update operations to modify existing documents
```
db.<collectionname>.update(
    { <query> },
    { $set: { <field1>: <value1>, <field2>: <value2>, ... }},
    } <options> }
)
```

Initial data
```
[
   { "_id": 1, "name": "Alice", "department": "HR", "position": "Assistant" },
   { "_id": 2, "name": "Bob", "department": "Sales", "position": "Salesman" }
]
```

update the position of an employee named "Alice" to "Manager"
```
db.employee.update(
    { name: "Alice" },
    { $set: { position: "Manager" }}
)
```

Updated data
```
[
   { "_id": 1, "name": "Alice", "department": "HR", "position": "Manager" },
   { "_id": 2, "name": "Bob", "department": "Sales", "position": "Salesman" }
]
```

##### $unset
The `$unset` operator removes a field from a document. It is used to delete a field by setting its value to `null`.
```
db.<collectionname>.update(
    { <query> }.
    { $unset: { <field1>: "", <field2>: "", ... } },
    { <options> }
)
```

remove the "position" field from "Bob"
```
db.employee.update(
    { name: "Bob" },
    { $unset: { position: "" } }
)
```

Updated data
```
[
   { "_id": 1, "name": "Alice", "department": "HR", "position": "Manager" },
   { "_id": 2, "name": "Bob", "department": "Sales" }
]
```

### updateOne
The `updateOne` method in MongoDB is used to update a single document in a collection that matches the specified filter criteria. This method updates only the first document that matches the filter criteria, even if there are multiple matching documents.
```
db.<collectionname>.updateOne(
    <filter>,
    <update>,
    {
        upsert: <boolean>,
        writeConcern: <document>,
        collation: <document>
    }
)
```
- filter: The selection criteria for the update. It determines which document to update.
- update: The modifications to apply.
- upsert: If true, performs an insert if no documents match the filter. Defaults to `false`.
- writeConcerns: The level of acknoledgement requested from MongoDB for write operations.
- collation: Specifies the collation to use for the operation.

Updating a single element
```
db.employee.updateOne(
    { name: "Alice" },
    { $set: { position: "Manager" } }
)
```

### updateMany
the `updateMany` method in MongoDB is used to update multiple documents in a collection that match the specified filter criteria. This method allows you to apply updates to all documents that math the filter.
```
db.<collectionname>.updateMany(

    }

```
- collation: specifies the collatoin to use for operation

Update all documents in the `employee` collection where the `department` is "Sales" to set their `status` to "active":
```

db.<collectionname>.findOneAndUpdate(
    <filter>,
    <update>,
    {
        projection: <document>,
        sort: <document>,
        upsert: <boolean>,
        returnNewDocument: <boolean>,
        maxTimeMS: <number>,
        collation: <document>,
        arrayFilters: [ <filterdocument1>, ...],
        bypassDocumentBalidation: <boolean>,
        writeConcern: <document>
    }
)
```
- projection: Specifies field to include or exclude
- sort: Specifies the order in which to return the documents
- returnNewDocument: If true, returns the updated document. Default to false, returns Original document.
- maxTimeMS: The maximum amount of time to run the query to run
- arrayFilters: Specifies which array elements to modify for an update operation on the array field
- bypassDocumentValidation: If true allows the write to opt out of document level validation.

Example
```
db.employee.findOneAndUpdate(
    { name: "Alice" },
    { $set: { position: "Manager" }},
    { returnNewDocument: true }
)
```

Advantages of `findOneAndUpdate` over `updateOne`
1. Returning the Document: either the updated or original document so you don't have to do another find operation on the database to verify that the document has been updated properly.
2. Atomic operation: the replacement approach of finding the document and replacing the document with updated value which requires making 2 calls can now be made with 1 call.

### findOneAndReplace
The `findOneAndReplace` method in MongoDB is used to find a single document that matches a filter, replace it with a new document, and return either the original or the replaced document. This method is useful whne you need to replace the entire document rather than updating specific fields.
```
db.<coleectionname>.findOneAndReplace(
    <filter>,
    <replacement>,
    {
        projection: <document>,
        sort: <document>,
        upsert: <boolean>,
        returnNewDocument: <boolean>,
        maxTimeMS: <number>,
        collation: <document>
    }
)
```
- replacement: The new document that will replace the existing document

By default `findOneAndReplace` returns the original document.

Replacing a Single Document and returning the new documetn
```
db.employee.findOneAndReplace(
   { name: "Alice" },
   { name: "Alice", department: "Finance", position: "CFO" },
   { returnNewDocument: true }
)
```

Use case for
- updateOne: Best for simple, partial updates where you don't need the updated document.
- findOneAndUpdate: Best for partial updates where you need the updated or original document immediately
- findOneAnndReplace: Best for scenarios where you need to completely replace a documetn and get the new or original document immediately.
